```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)

# Convert 'Date' column to datetime format if it's not already
df['DATE'] = pd.to_datetime(df['DATE'])

# Verify data types
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt

df_tunisia = df[df['COUNTRY'] == 'Tunisia']
df_cameroon = df[df['COUNTRY'] == 'Cameroon']

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df_tunisia['DATE'], df_tunisia['TAVG'], label='Tunisia')
plt.plot(df_cameroon['DATE'], df_cameroon['TAVG'], label='Cameroon')
plt.xlabel('DATE')
plt.ylabel('TAVG (F)')
plt.title('Average Temperature Fluctuations in Tunisia and Cameroon')
plt.legend()
plt.show()
```

```python
start_date = '1980-01-01'
end_date = '2005-12-31'

df_tunisia_zoomed = df_tunisia[(df_tunisia['DATE'] >= start_date) & (df_tunisia['DATE'] <= end_date)]
df_cameroon_zoomed = df_cameroon[(df_cameroon['DATE'] >= start_date) & (df_cameroon['DATE'] <= end_date)]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_tunisia_zoomed['DATE'], df_tunisia_zoomed['TAVG'], label='Tunisia')
plt.plot(df_cameroon_zoomed['DATE'], df_cameroon_zoomed['TAVG'], label='Cameroon')
plt.xlabel('Year')
plt.ylabel('Average Temperature (F)')
plt.title('Average Temperature Fluctuations in Tunisia and Cameroon (1980-2005)')
plt.legend()
plt.show()
```

```python
df_senegal = df[df['COUNTRY'] == 'Senegal']

senegal_1980_2000 = df_senegal[(df_senegal['DATE'] >= '1980-01-01') & (df_senegal['DATE'] <= '2000-12-31')]
senegal_2000_2023 = df_senegal[(df_senegal['DATE'] >= '2000-01-01') & (df_senegal['DATE'] <= '2023-12-31')]

plt.figure(figsize=(14, 7))
plt.hist(senegal_1980_2000['TAVG'], bins=30, alpha=0.5, label='1980-2000')
plt.hist(senegal_2000_2023['TAVG'], bins=30, alpha=0.5, label='2000-2023')
plt.xlabel('Average Temperature (F)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution in Senegal (1980-2000 vs 2000-2023)')
plt.legend()
plt.show()
```

```python
avg_temp_per_country = df.groupby('COUNTRY')['TAVG'].mean().reset_index()
```
